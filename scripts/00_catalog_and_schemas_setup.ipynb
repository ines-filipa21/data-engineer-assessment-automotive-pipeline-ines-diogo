# Catalog and Schemas Setup

## Objective
This notebook creates the catalog and schemas used by the automotive medallion architecture.

## Main goals
- create the project catalog
- create the bronze, silver, and gold schemas

## Notes
This setup is executed once before running the bronze, silver, and gold layer notebooks.

```python
catalog_name = "automotive"

bronze_schema = "bronze"
silver_schema = "silver"
gold_schema = "gold"
```

## Create catalog

Create the main catalog used by the project if it does not already exist.

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog_name}")
```

## Create schemas

Create the bronze, silver, and gold schemas used in the medallion architecture.

```python
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{bronze_schema}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{silver_schema}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{gold_schema}")
```

## Final note

The catalog and schemas are now ready to support bronze, silver, and gold table creation.
